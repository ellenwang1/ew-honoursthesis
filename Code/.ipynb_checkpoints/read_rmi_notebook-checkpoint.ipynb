```python
name = nib.load(r'C:\Users\ellen\Documents\ew-honoursthesis\Data\forAudrey.tar\T1softTissMask_withLacunes_35.tar\0046_T1softTiss.nii.gz')
data = name.get_fdata()
data_aff  = name.affine
data_hdr  = name.header
data_shape = name.shape
print(data_shape)
```

```python
def show_slices(slices):
   fig, axes = plt.subplots(1, len(slices))
   for i, slice in enumerate(slices):
       axes[i].imshow(slice.T, cmap="gray", origin="lower")
slice_0 = data[128, :, :]
slice_1 = data[:, 128, :]
slice_2 = data[:, :, 130]
show_slices([slice_0, slice_1, slice_2])
plt.suptitle("Center slices for MRI image") 
print(slice_1[0])
```

```python
for i in range(0, 256, 64):
    for j in range(0, 256, 64):
        for k in range(0, 190, 38):
            slice_0 = data[i, :, :]
            slice_1 = data[:, j, :]
            slice_2 = data[:, :, k]
            show_slices([slice_0, slice_1, slice_2])
            plt.suptitle("Center slices for MRI image") 
```
